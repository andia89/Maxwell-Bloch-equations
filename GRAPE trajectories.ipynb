```python
from qutip import *
from pylab import *
from scipy.io import loadmat
from scipy.interpolate import interp1d
import scipy as sp
import numpy as np

def squeeze_pulse(pulse, squeeze_factor):
    '''Expands or contracts a pulse, keeping the integral under the curve constant. Squeeze_factor < 1 makes pulse shorter.'''
    new_len = round(np.shape(pulse)[0] * squeeze_factor)
    time_step = pulse[1,0] - pulse[0,0]
    p = np.zeros((new_len, 3))
    p[:,0] = np.arange(0., new_len*time_step, time_step)

    #old_area = sp.integrate.trapz(pulse[:,1], dx=time_step)

    f1 = sp.interpolate.interp1d(pulse[:,0]*squeeze_factor, pulse[:,1])
    f2 = sp.interpolate.interp1d(pulse[:,0]*squeeze_factor, pulse[:,2])

    p[:,1] = f1(p[:,0])/squeeze_factor
    p[:,2] = f2(p[:,0])/squeeze_factor

    #new_area = sp.integrate.trapz(p[:,1], dx=time_step)
    
    return p


def pulse_rwa(t, args):
    p=args
    idx = searchsorted(p[:,0], t*1e3)
    #if t is slightly larger than some time in list, take the y-value to the left:
    if idx==len(p) or (p[idx,0] != t*1e3 and idx != 0):
        idx = idx - 1 
    if t*1e3>p[-1,0]:
        return 0  
    return p[idx, 1] 

def pulse_rwa_q(t, args):
    p=args
    idx = searchsorted(p[:,0], t*1e3)
    #if t is slightly larger than some time in list, take the y-value to the left:
    if idx==len(p) or (p[idx,0] != t*1e3 and idx != 0):
        idx = idx - 1
    if t*1e3>p[-1,0]:
        return 0
    return p[idx, 2]

def rolling_window(a, window):
    """
    Make an ndarray with a rolling window of the last dimension

    Parameters
    ----------
    a : array_like
        Array to add rolling window to
    window : int
        Size of rolling window

    Returns
    -------
    Array that is a view of the original array with a added dimension
    of size w.

    Examples
    --------
    >>> x=np.arange(10).reshape((2,5))
    >>> rolling_window(x, 3)
    array([[[0, 1, 2], [1, 2, 3], [2, 3, 4]],
           [[5, 6, 7], [6, 7, 8], [7, 8, 9]]])

    Calculate rolling mean of last dimension:
    >>> np.mean(rolling_window(x, 3), -1)
    array([[ 1.,  2.,  3.],
           [ 6.,  7.,  8.]])

    """
    if window < 1:
        raise ValueError, "`window` must be at least 1."
    if window > a.shape[-1]:
        raise ValueError, "`window` is too long."
    shape = a.shape[:-1] + (a.shape[-1] - window + 1, window)
    strides = a.strides + (a.strides[-1],)
    return np.lib.stride_tricks.as_strided(a, shape=shape, strides=strides)
```

```python
#generate random sample from ornstein-uhlenbeck process
def simulate_ornstein_uhlenbeck(S0, mu, sigma, l, deltat, t):
    periods = floor(t/deltat)
    S = zeros(periods)
    S[0] = S0
    dWt = sqrt(deltat) * randn(periods,1)
    for t in range(1, int(periods)):
        dSt = l*(mu-S[t-1])*deltat + sigma*dWt[t]
        S[t] = S[t-1] + dSt
    return S
```
